### Hosotani solution for 2-flavor Schwinger model

```python
γ = 0.5772156649 # Euler-Mascheroni constant
```

```python
# input parameters:
β = 4.0
L = 10 # N_t
```

$\mu = \sqrt{\frac{2}{\pi\,\beta}}$

```python
μ = n(sqrt(2 / (pi * β))) # eta-mass
```

```python
μ
```

```python
μ * L # this should be >> 1
```

```python
1.0 / (2.0 * L * sqrt(μ * L)) # solution is valid for m << than this number 
```

```python
k = n(4 * sqrt(2) * sqrt(μ * L * exp(γ) / (4 * pi))) # slope for small m
```

```python
k
```

```python
var('m')
plot([k * m, (4 * exp(2 * γ) * μ * m^2)^(1/3)], 0.0001, 0.23, gridlines = True)
```
