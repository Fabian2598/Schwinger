# FiniteT

### Hosotani solution for 2-flavor Schwinger model

```python
γ = 0.5772156649 # Euler-Mascheroni constant
```

```python
# input parameters:
β = 4.0
L = 20 # N_t
```

$\mu = \sqrt{\frac{2}{\pi\,\beta}}$

```python
μ = n(sqrt(2 / (pi * β))) # eta-mass
g = 1.0 / sqrt(β)
```

```python
μ
```

```python
g
```

```python
μ * L # this should be >> 1
```

## Pion mass

```python
b = 1.0 / (2.0 * L * sqrt(μ * L)) # solution is valid for m << than this number 
```

```python
b
```

```python
k = n(4 * sqrt(2) * sqrt(μ * L * exp(γ) / (4 * pi))) # slope for small m
```

```python
k
```

```python
var('m')
plot([k * m, (4 * exp(2 * γ) * μ * m^2)^(1/3)], 0.0001, 0.23,
  gridlines = True, axes_labels = [r"$m$", r"$m_\pi$"]) + line(
  [[b, 0], [b, 0.2]], linestyle = ":", color = 'red')
```

## Condensate $<\bar{\psi}\psi>$

```python
b = 1.0 / (L * sqrt(μ * L)) # solution is valid for m << than this number 
```

```python
b
```

```python
k = n(2 * exp(γ) * μ * L / pi^2) # slope for small m
```

```python
k
```

```python
var('m')
plot([k * m, (exp(4 * γ) * m * μ^2 / (4 * pi^3))^(1/3)],
  0.0001, 0.1, gridlines = True, axes_labels=[r'$m$', r'$<\bar{\psi}\psi>$']) + line(
  [[b, 0], [b, 0.1]], linestyle = ":", color = 'red')
```

```python
var('x') # x = m / g  =>  m = x * g
plot([k * (x * g) / g, (exp(4 * γ) * (x * g) * μ^2 / (4 * pi^3))^(1/3) / g], 0.0001, 0.1 / g,
  gridlines = True, axes_labels=[r'$m/g$', r'$<\bar{\psi}\psi>/g$']) + line(
  [[b / g, 0], [b / g, 0.1 / g]], linestyle = ":", color = 'red')
```
