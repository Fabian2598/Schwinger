### Besselove funkcije

- ovo, nažalost, ne zna integrirati - trebala bi biti Besselova funkcija `bessel_I(0, b)`

> $I_0(\beta) \equiv \frac{1}{2 \pi} \int_{-\pi}^{\pi} d\theta \, e^{\beta \cos(\theta)}$

```python
assume(b > 0)
integral(exp(b * cos(t)), t, -pi, pi)
```

```python
tocke = [[b, numerical_integral(exp(b * cos(t)) / (2 * pi), -pi, pi)[0]] for b in numpy.arange(0, 2.1, 0.125)]
```

```python
list_plot(tocke) + plot(bessel_I(0, x), 0, 2)
```

***
- ni ovo, nažalost, ne zna integrirati - trebala bi biti Besselova funkcija `bessel_I(1, b)`

> $I_1(\beta) \equiv \frac{1}{2 \pi} \int_{-\pi}^{\pi} d\theta \, \cos(\theta) \, e^{\beta \cos(\theta)}$

- OPREZ! - ovo ga zablesira, treba napraviti interrupt kernela da se izađe...

```python
assume(b > 0)
integral(cos(t) * exp(b * cos(t)), t, -pi, pi)
```

- i to ćemo provjeriti numeričkom integracijom

```python
tocke = [[b, numerical_integral(cos(t) * exp(b * cos(t)) / (2 * pi), -pi, pi)[0]] for b in numpy.arange(0, 2.1, 0.125)]
```

```python
list_plot(tocke) + plot(bessel_I(1, x), 0, 2)
```

***
- usporedba dviju Besselovih funkcija:

```python
plot(bessel_I(0, x), 0, 2) + plot(bessel_I(1, x), 0, 2, color = 'red')
```

### Baždarno djelovanje (engl. *gauge action*)

```python
plot(x * (1 - bessel_I(1, x) / bessel_I(0, x)), 0, 5)
```

```python
[[b, b * (1.0 - bessel_I(1, RealNumber(b)) / bessel_I(0, RealNumber(b)))] for b in numpy.arange(0, 5.25, 0.25)]
```

- srednja vrijednost plaketa

```python
[[b, bessel_I(1, RealNumber(b)) / bessel_I(0, RealNumber(b))] for b in numpy.arange(0, 5.25, 0.25)]
```

```python
tocke = [[b, bessel_I(1, RealNumber(b)) / bessel_I(0, RealNumber(b))] for b in numpy.arange(0, 5.25, 0.25)]
```

```python
list_plot(tocke, ymax = 1) + point((4.0, 0.86279), color = 'yellow')
```

- dobro se slaže s vrijednošću plakete iz članka Duerr & Hoelbling, Phys.Rev.D **69**, 034503 (2004): 0.86279 za $\beta = 4.0$ (žuta točka na gornjem grafu)
