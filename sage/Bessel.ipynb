- check that `bessel_I(0, x)` in Sagemath is `scipy.special.i0(x)` in Python

```python
var('x')
sage = plot(bessel_I(0, x), 0, 3, color = 'red', ymin = 0)

import scipy.special as sp
b = [[x, sp.i0(x)] for x in srange(0, 3, 0.1)]

python = list_plot(b)
show(sage + python)
```
